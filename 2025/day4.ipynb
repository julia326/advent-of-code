```python
import io
import pandas as pd
```

```python
example_input = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@."""
```

```python
def make_df(lines):
    return pd.DataFrame([list(line) for line in lines])

df = make_df(example_input.split('\n'))
df
```

```python
def get_adjacent_pos_list(df, i, j):
    pos_list = []
    for delta_i in [-1, 0, 1]:
        new_i = i + delta_i
        if new_i < 0 or new_i >= df.shape[0]:
            continue
        for delta_j in [-1, 0, 1]:
            new_j = j + delta_j
            if new_j < 0 or new_j >= df.shape[1] or (new_i == i and new_j == j):
                continue
            pos_list.append([new_i, new_j])
    return pos_list

def get_adjacent_rolls_count(df, i, j):
    num_adjacent_rolls = 0
    for [i2, j2] in get_adjacent_pos_list(df, i, j):
        if df.iloc[i2, j2] == '@':
            num_adjacent_rolls += 1
    return num_adjacent_rolls

def part1(lines):
    df = make_df(lines)
    print(df.shape)
    num_reachable_rolls = 0
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            val = df.iloc[i, j]
            if val != '@':
                continue
            if get_adjacent_rolls_count(df, i, j) < 4:
                num_reachable_rolls += 1
    return num_reachable_rolls
        
part1(example_input.split('\n'))
```

```python
my_input = open('day4.txt').read().split('\n')[:-1]
part1(my_input)
```

```python
# lines = example_input.split('\n')
lines = my_input

df = make_df(lines)
print(df.shape)

new_df = df.copy()
num_reachable_rolls = 0

while True:
    print(f'Current num reachable rolls: {num_reachable_rolls}')
    any_updates = False
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            val = df.iloc[i, j]
            if val != '@':
                continue
            if get_adjacent_rolls_count(df, i, j) < 4:
                num_reachable_rolls += 1
                new_df.iloc[i, j] = '.'  # remove the roll
                any_updates = True
    if not any_updates:
        break
    df = new_df
```
