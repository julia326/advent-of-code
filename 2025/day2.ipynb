```python
def is_invalid(id_str):
    half_len = int(len(id_str)/2)
    first_half = id_str[:half_len]
    second_half = id_str[half_len:]
    return first_half == second_half

is_invalid('11')
```

```python
def all_factors(num):
    factors = []
    for i in range(1, num):
        if num % i == 0:
            factors.append(i)
    return factors

def is_invalid2(num):
    num_str = str(num)
    factors = all_factors(len(num_str))
    for factor in factors:
        # this is the length a repeat is allowed to be
        repeat = num_str[:factor]
        multiplier = int(len(num_str)/factor)
        if num_str == repeat * multiplier:
            return True
    return False
```
