```python
example_input = """987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
for line in example_input.split('\n'):
    print(line)
```

```python
def part1(lines):
    max_num_sum = 0
    for line in lines:
        max_num = 0
        for i in range(len(line)):
            for j in range(i, len(line)):
                if i == j:
                    continue
                num = int(line[i])*10 + int(line[j])
                if num > max_num:
                    max_num = num
        max_num_sum += max_num
    return max_num_sum

part1(example_input.split('\n'))
```

```python
my_input = open('day3.txt').read().split('\n')[:-1]
part1(my_input)
```

```python
def find_max_num(line):
    max_digits = []
    max_i = -1

    for j in range(12):
        # find the max number between the last highest number and enough buffer
        # before the end of the array to make a 12-digit number overall
        start_i = max_i + 1
        end_i = len(line) - (12 - j)
        max_digit = 0
        max_i = None
        for i in range(start_i, end_i + 1):
            if int(line[i]) > max_digit:
                max_digit = int(line[i])
                max_i = i
#         print(f'After {j}th round: new digit {max_digit} at position {i}')
        max_digits.append(max_digit)

    return int(''.join([str(x) for x in max_digits]))

find_max_num('818181911112111')
```

```python
def part2(lines):
    return sum(find_max_num(line) for line in lines)

part2(example_input.split('\n'))
```
