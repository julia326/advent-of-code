```python
cur_pos = 50

def next_pos(cur_pos, move):
    direction = move[0]
    num = int(move[1:])
    if move[0] == 'L':
        return (cur_pos - num) % 100
    else:
        return (cur_pos + num) % 100
    
def part1(moves):
    cur_pos = 50
    num_0 = 0
    for move in moves:
        cur_pos = next_pos(cur_pos, move)
        if cur_pos == 0:
            num_0 += 1
    return num_0

part1(example_input.split('\n'))
```

```python
import math

def part2(moves, start_pos=50):
    cur_pos = start_pos
    num_0 = 0
    for move in moves:
        print(f'Cur_pos: {cur_pos}, now facing move: {move}')
        direction = move[0]
        num = int(move[1:])
        if move[0] == 'L':
            next_cur_pos = (cur_pos - num) % 100
            if cur_pos - num <= 0:  # we went around; question is, how many times?
                if next_cur_pos == 0:
                    num_0 += 1
                else:
                    num_rotations = math.ceil(abs(cur_pos - num) / 100)
                    if cur_pos == 0:
                        num_rotations -= 1
                    num_0 += num_rotations
            cur_pos = next_cur_pos
        else:
            next_cur_pos = (cur_pos + num) % 100
            if cur_pos + num >= 100:  # we went around some more
                if next_cur_pos == 0:
                    num_0 += 1
                else:
                    num_rotations = math.floor((cur_pos + num) / 100)
                    num_0 += num_rotations
            cur_pos = next_cur_pos
        print(f'At end of loop, cur_pos: {cur_pos}, with num_0: {num_0}')
    return num_0

# part2(example_input.split('\n'))
part2(['L808'], start_pos=8)
# part2(['L50'])
```
