```python
example_input = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""
```

```python
def is_fresh(num, range_list):
    for r in range_list:
        if num >= r[0] and num <= r[1]:
            return True
    return False

def part1(input_str):
    ranges, nums = input_str.split('\n\n')
    range_list = []
    for r in ranges.split('\n'):
        range_list.append([int(x) for x in r.split('-')])
    num_fresh = 0
    for num in nums.split('\n'):
        if num == '':
            break
        num = int(num)
        if is_fresh(num, range_list):
            num_fresh += 1
    return num_fresh

part1(example_input)
```

```python
def part2(input_str):
    all_nums = set()
    ranges = input_str.split('\n\n')[0].split('\n')
    print(f'{len(ranges)} ranges')
    for r in ranges:
        start, end = [int(x) for x in r.split('-')]
        start = int(start)
        end = int(end) + 1
        all_nums.update(list(range(start, end)))
        print(f'{len(all_nums)} total nums so far')
    return len(all_nums)

part2(example_input)
```

```python
"""
iterate from the beginning. Say we're looking at two ranges and trying to figure out total coverage
between the two. A few scenarios:
1
------
----------

2
------
  -------

3
------
  --
 
4
-----
        ------

As we process the ranges, we want to keep track of the "merged" ranges over time. So as we go down
the list, we update a new list of ranges, always modifying at most the last range. So we start with the first
range which is:
------

Scenario 1: modify last range, extend endpoint
Scenario 2: modify last range, extend endpoint
Scenario 3: change nothing
Scenario 4: add new range
"""
```

```python
# parse, sort range list by start
def make_sorted_range_list(input_str):
    ranges = input_str.split('\n\n')[0].split('\n')
    range_list = []
    for r in ranges:
        start, end = [int(x) for x in r.split('-')]
        range_list.append((start, end))
    sorted_range_list = sorted(range_list, key=lambda x: x[0])
    return sorted_range_list

def compute_new_ranges(sorted_range_list):
    new_ranges = [sorted_range_list[0]]
    for (start, end) in sorted_range_list:
#         print(f'Considering {start}-{end}')
        (last_start, last_end) = new_ranges[-1]
        # scenario 3: no change needed
        if end < last_end:  
            continue
        # scenarios 1, 2: extend last range's end
        elif start <= last_end and end > last_end:
            new_ranges[-1] = [last_start, end]
        # scenario 4: new range
        elif start > last_end:
            new_ranges.append([start, end])
#         print(new_ranges)
#         print('\n')

    return new_ranges

def sum_ranges(ranges):
    # sum the ranges in new_ranges
    range_sum = 0
    for (start, end) in ranges:
        range_sum += (end - start + 1)

    return range_sum

sorted_range_list = make_sorted_range_list(example_input)
new_ranges = compute_new_ranges(sorted_range_list)
sum_ranges(new_ranges)
```
