```python
def get_next_dir(cur_dir):
    next_dir_map = {'^': '>', '>': 'v', 'v': '<', '<': '^'}
    return next_dir_map[cur_dir]

# find starting position
def get_start_pos_and_dir(df):
    cur_dir = None
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if df.iloc[i, j] in ['^', '>', 'v', '<']:
                cur_dir = df.iloc[i, j]
                return ([i, j], cur_dir)
            
def part1(input_str):
    # parse df
    lines = []
    for line in input_str.split('\n'):
        lines.append(list(line))

    df = pd.DataFrame(lines)
    [i, j], cur_dir = get_start_pos_and_dir(df)

    print('%d, %d' % (i, j))
    all_pos = [(i, j)]
    df.iloc[i, j] = '.'

    while i >= 0 and i < df.shape[0] and j >= 0 and j < df.shape[1]:
        # if the current position is a snag, reset to the last known position and change the direction
        if df.iloc[i, j] == '#':
            i, j = all_pos[-1]
            cur_dir = get_next_dir(cur_dir)
            continue

        all_pos.append((i, j))
        # print('%d, %d' % (i, j))
        cur_value = df.iloc[i, j]
        assert cur_value == '.'

        if cur_dir == '^':  # up
            i -= 1
        elif cur_dir == 'v':  # down
            i += 1
        elif cur_dir == '>':  # right
            j += 1
        elif cur_dir == '<':  # left
            j -= 1
        else:
            raise ValueError('wtf')

    return set(all_pos)
    
len(part1(example_input))
```
