```python
example_input = '125 17'

def blink(input_lst):  # list of strings
    new_lst = []
    for num_str in input_lst:
        if num_str == '0':
            new_lst.append('1')
        elif len(num_str) % 2 == 0:
            midpoint = int(len(num_str)/2)
            num1 = int(num_str[:midpoint])
            num2 = int(num_str[midpoint:])
            new_lst.append(str(num1))
            new_lst.append(str(num2))
        else:
            new_lst.append(str(2024 * int(num_str)))
    return new_lst

input_lst = example_input.split(' ')
for i in range(25):
    input_lst = blink(input_lst)

len(input_lst)
```

```python
def convert(num, num_blinks):
    input_lst = [num]
    for i in range(num_blinks):
        print(i)
        input_lst = blink(input_lst)
    return input_lst

convert('6', 75-34)
```

```python
memoized = {}  # (num, epochs) -> list(nums)

def convert(num, epochs):
    print('Calling with %s, %d' % (num, epochs))
    if (num, epochs) in memoized:
        return memoized[(num, epochs)]
    
    conversion = blink([num])
    if epochs == 1:
        memoized[(num, 1)] = conversion
        return conversion
    
    else:
        full_ans = []
        for other_num in conversion:
            ans = convert(other_num, epochs-1)
            print('Ans:')
            print(ans)
            full_ans.extend(ans)
            memoized[(num, epochs)] = full_ans
    
out = convert('2', 4)

memoized
```
