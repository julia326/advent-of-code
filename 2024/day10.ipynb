```python
example_input = """89010123
78121874
87430965
96549874
45678903
32019012
01329801
10456732"""

import pandas as pd

def parse_input_to_df(input_str):
    lines = []
    for line in input_str.split('\n'):
        lines.append(list(line))

    return pd.DataFrame(lines)

df = parse_input_to_df(example_input)
df
```

```python
import networkx as nx

def in_range(df, i, j):
    return i >= 0 and i < df.shape[0] and j >= 0 and j < df.shape[1]

def make_tree(df, i, j, graph):
    # check in all 4 directions
    new_positions = [
        (i-1, j),  # up
        (i+1, j),  # down
        (i, j-1),  # left
        (i, j+1)   # right
    ]
    
    for (i2, j2) in new_positions:
        if in_range(df, i2, j2) and int(df.iloc[i2, j2]) == int(df.iloc[i, j]) + 1:
            graph.add_edge((i, j), (i2, j2))
            make_tree(df, i2, j2, graph)
            
graph = nx.DiGraph()
make_tree(df, 0, 2, graph)

graph.edges()
```

```python
def part1(df):
    pos0s = []
    pos9s = []
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if df.iloc[i, j] == '0':
                pos0s.append((i, j))
            elif df.iloc[i, j] == '9':
                pos9s.append((i, j))

    total = 0
    for pos0 in pos0s:         
        graph = nx.DiGraph()
        make_tree(df, pos0[0], pos0[1], graph)
        count = 0
        for pos9 in pos9s:
            if graph.has_node(pos9) and nx.has_path(graph, pos0, pos9):
                count += 1
        total += count

    print(total)
    
part1(df)
```

```python
# part 2: get the number of distinct trails from each trailhead
df = parse_input_to_df(a)

pos0s = []
pos9s = []
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        if df.iloc[i, j] == '0':
            pos0s.append((i, j))
        if df.iloc[i, j] == '9':
            pos9s.append((i, j))

total = 0
for pos0 in pos0s:         
    graph = nx.DiGraph()
    make_tree(df, pos0[0], pos0[1], graph)
    count = 0
    for pos9 in pos9s:
        if graph.has_node(pos9):
            count += len(list(nx.all_simple_paths(graph, pos0, pos9)))
    total += count

print(total)
```
