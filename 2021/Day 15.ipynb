```python
def get_risk(df, cur_x, cur_y, memoized):
    if cur_x == df.shape[0] - 1 and cur_y == df.shape[1] - 1:
        # base case, reached the end
        return 0
    else:
        if (cur_x, cur_y) in memoized:
            return memoized[(cur_x, cur_y)]
        # get down and right risk if not out of bounds
        cur_value = df.iloc[cur_x, cur_y]
        right_risk = cur_value + get_risk(df, cur_x + 1, cur_y, memoized) \
            if cur_x < df.shape[0] - 1 else sys.maxsize
        down_risk = cur_value + get_risk(df, cur_x, cur_y + 1, memoized) \
            if cur_y < df.shape[1] - 1 else sys.maxsize

        risk = min(right_risk, down_risk)
        memoized[(cur_x, cur_y)] = risk
        return risk

memoized = {}  # maps position to risk, to not recompute every time
df = load_example()
get_risk(df, 0, 0, memoized)
```

```python
def upgrade(val, i):
    val += i
    return val if val <= 9 else val - 9

def upgrade_df(df, i):
    return df.applymap(lambda x: upgrade(x, i))

def make_larger_cave(df):
    df_rows = []
    for i in range(5):
        df_row_list = [upgrade_df(df, i)]
        for j in range(4):
            df_row_list.append(upgrade_df(df, i+j+1))
        df_row = pd.concat(df_row_list, axis=1)
        df_rows.append(df_row)

    new_df = pd.concat(df_rows, axis=0)
    new_df.index = list(range(new_df.shape[0]))
    new_df.columns = list(range(new_df.shape[1]))
    return new_df

new_df = make_larger_cave(load_example())
new_df
```

```python
def get_risk(df, cur_x, cur_y, memoized, memoized_directions):
    if cur_x == df.shape[0] - 1 and cur_y == df.shape[1] - 1:
        memoized[(cur_x, cur_y)] = 0
        # base case, reached the end
        return 0
    else:
        if (cur_x, cur_y) in memoized:
            return memoized[(cur_x, cur_y)]
        # get down and right risk if not out of bounds
        down_risk = df.iloc[cur_x+1, cur_y] + get_risk(df, cur_x+1, cur_y, memoized, memoized_directions) \
            if cur_x < df.shape[0]-1 else sys.maxsize
        right_risk = df.iloc[cur_x, cur_y+1] + get_risk(df, cur_x, cur_y+1, memoized, memoized_directions) \
            if cur_y < df.shape[1]-1 else sys.maxsize

        risk = min(down_risk, right_risk)
        memoized[(cur_x, cur_y)] = risk
        memoized_directions[(cur_x, cur_y)] = 'DOWN' if down_risk < right_risk else 'RIGHT'
            
        return risk

memoized = {}  # maps position to risk, to not recompute every time
memoized_directions = {}
df = load_example()
risk = get_risk(df, 0, 0, memoized, memoized_directions)
risk
```

```python
def is_in_bounds(x, y, df):
    return x >= 0 and x < df.shape[0] and y >= 0 and y < df.shape[1]

def get_risk_part2(df, cur_x, cur_y, memoized, path_so_far):
    if cur_x == df.shape[0] - 1 and cur_y == df.shape[1] - 1:
        memoized[(cur_x, cur_y)] = 0
        # base case, reached the end
        return 0
    else:
        if (cur_x, cur_y) in memoized:
            return memoized[(cur_x, cur_y)]
        
        # consider all other paths
        next_pos_list = [
            (cur_x+1, cur_y),
            (cur_x, cur_y+1),
            (cur_x-1, cur_y),
        ]
        risks = []
        for (next_x, next_y) in next_pos_list:
            path_so_far.append((cur_x, cur_y))
            if is_in_bounds(next_x, next_y, df) and not (next_x, next_y) in path_so_far:
                next_risk = df.iloc[next_x, next_y] + get_risk_part2(
                    df, next_x, next_y, memoized, path_so_far)
                risks.append(next_risk)
            path_so_far.pop()
                
        if len(risks) > 0:  # ended up somewhere unfriendly, return max
            risk = min(risks)
            memoized[(cur_x, cur_y)] = risk
            return risk
        
        else:
            return 0

memoized = {}  # maps position to risk, to not recompute every time
path_so_far = []  # to make sure there are no cycles
df = load_example()
risk = get_risk_part2(df, 0, 0, memoized, path_so_far)
risk
```
