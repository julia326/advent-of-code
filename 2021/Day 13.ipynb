```python
example = """
6,10
0,14
9,10
0,3
10,4
4,11
6,0
6,12
4,1
0,13
10,12
3,4
3,0
8,4
1,10
2,14
8,10
9,0

fold along y=7
fold along x=5
"""

points, folds = example.split('\n\n')
points = points.split('\n')[1:]
folds = folds.split('\n')[:-1]
```

```python
def make_df(points):
    max_x = 0
    max_y = 0
    for point in points:
        x, y = point.split(',')
        max_x = max(max_x, int(x))
        max_y = max(max_y, int(y))

    df = pd.DataFrame(index=range(max_y+1), columns=range(max_x+1)).fillna('.')
    
    for point in points:
        x, y = point.split(',')
        df.iloc[int(y), int(x)] = '#'
    return df
        
df = make_df(points)
df
```

```python
def get_bottom_half(df, index_to_fold):
    return df.iloc[index_to_fold+1:,:].copy()

def get_top_half(df, index_to_fold):
    return df.iloc[:index_to_fold,:].copy()

def flip_vertically(df):
    new_df = df.iloc[::-1]
    new_df.index = list(range(new_df.shape[0]))
    return new_df

flip_vertically(get_bottom_half(df, 7))
```

```python
get_top_half(df, 7)
```

```python
def fold_at_y(df, num):
    top_half = get_top_half(df, num).replace({'#': True, '.': False})
    bottom_half = flip_vertically(get_bottom_half(df, num)).replace({'#': True, '.': False})
    new_df = top_half | bottom_half
    return new_df.replace({True: '#', False: '.'})
    
new_df = fold_at_y(df, 7)
new_df
```

```python
def fold_at_x(df, num):
    left_half = get_left_half(df, num).replace({'#': True, '.': False})
    right_half = flip_horizontally(get_right_half(df, num)).replace({'#': True, '.': False})
    new_df = left_half | right_half
    return new_df.replace({True: '#', False: '.'})

final_df = fold_at_x(new_df, 5)
final_df
```
