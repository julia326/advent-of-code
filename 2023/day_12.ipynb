# Part 1

```python
example = """???.### 1,1,3
.??..??...?##. 1,1,3
?#?#?#?#?#?#?#? 1,3,1,6
????.#...#... 4,1,1
????.######..#####. 1,6,5
?###???????? 3,2,1"""
```

```python
line = '???.### 1,1,3'

"""
how many versions of . and # correspond to 1,1,3 counts?

#.#.### 1,1,3

for the line, generate all possible strings, and figure out the consecutive counts for each
"""

def count_damage(record):
    counts = []
    in_stretch = False
    for char in record:
        if char == '#':
            if in_stretch:
                counts[-1] += 1
            else:
                counts.append(1)
                in_stretch = True
        else:
            in_stretch = False

    return counts

def convert_to_int_array(counts):
    # counts is a string like '1,1,3'
    return [int(x) for x in counts.split(',')]
```

```python
def generate_all_strings(record, out, word_list):
    # base case: we're at the end of the word, and "out" is a single string record
    if record == '':
        word_list.append(out)
    # recursive case: "out" keeps growing the word
    elif record[0] == '.':
        generate_all_strings(record[1:], out + '.', word_list)
    elif record[0] == '#':
        generate_all_strings(record[1:], out + '#', word_list)
    else:
        generate_all_strings(record[1:], out + '.', word_list)
        generate_all_strings(record[1:], out + '#', word_list)
        
word_list = []
record = '???.###'
generate_all_strings(record, '', word_list)
for word in word_list:
    print('%s: %s' % (word, str(count_damage(word))))
```

```python
def process_line(line='???.### 1,1,3'):
    num_hits = 0
    record, counts = line.split(' ')
    word_list = []
    generate_all_strings(record, '', word_list)
    for word in word_list:
        if count_damage(word) == convert_to_int_array(counts):
            num_hits += 1
    return num_hits
            
total = 0
for line in example.split('\n'):
    total += process_line(line)

total
```

```python
line = '.# 1'

def unfold(line):
    record, counts = line.split(' ')
    new_record = '?'.join([record]*5)
    new_counts = ','.join([counts]*5)
    return (new_record, new_counts)

def process_line_part2(line):
    num_hits = 0
    record, counts = unfold(line)
    word_list = []
    generate_all_strings(record, '', word_list)
    int_counts = convert_to_int_array(counts)
    for word in word_list:
        if count_damage(word) == int_counts:
            num_hits += 1
    return num_hits

for line in example.split('\n'):
    print(process_line_part2(line))
```

```python
def resolve_record(line):
    candidate_strs = []
    pos = [i for i in range(len(line)) if line[i] == '?']
    for i in range(pow(2, len(pos))):
        line_lst = list(line)
        bin_rep = bin(i)[2:]
        out = '0' * (len(pos) - len(bin_rep)) + bin_rep
        for i, char in enumerate(out):
            if char == '0':
                line_lst[pos[i]] = '.'
            else:
                line_lst[pos[i]] = '#'
        candidate_strs.append(''.join(line_lst))
    return candidate_strs

resolve_record('???.###')
```

```python
def process_line(line='???.### 1,1,3'):
    num_hits = 0
#     record, counts = line.split(' ')
    record, counts = unfold(line)
    word_list = resolve_record(record)
    for word in word_list:
        if count_damage(word) == convert_to_int_array(counts):
            num_hits += 1
    return num_hits
            
total = 0
for line in example.split('\n'):
    total += process_line(line)

total
```
